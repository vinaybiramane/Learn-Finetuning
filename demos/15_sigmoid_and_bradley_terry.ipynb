# 15 · Modeling "A beats B"

**Recap (13–14):** we can give a sequence a **log-probability**, and we can build
**(chosen, rejected)** pairs for free. Now we need to turn *"chosen is better"* into
a number we can optimize. Two small tools do it: the **sigmoid** and the
**Bradley-Terry** model.

```python
import numpy as np
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

## Step 1 — The sigmoid squashes any number into a probability

$$\sigma(z) = \frac{1}{1 + e^{-z}}$$

It maps any real number into `(0, 1)`: very negative → near 0, zero → exactly 0.5,
very positive → near 1. That's how we'll express "how confident are we that chosen
beats rejected."

```python
for z in [-3, -1, 0, 1, 3]:
    print(f"sigmoid({z:+d}) = {sigmoid(z):.2f}")
```

## Step 2 — Bradley-Terry: probability that the better one wins

Give each option a **score**. The Bradley-Terry model says the probability that
*chosen* beats *rejected* depends only on the **gap** between their scores:

$$P(\text{chosen} \succ \text{rejected}) = \sigma(\text{score}_\text{chosen} - \text{score}_\text{rejected})$$

Equal scores → 0.5 (a toss-up). The bigger chosen's lead, the closer to 1.

```python
for gap in [-2, -1, 0, 1, 2, 4]:
    print(f"score gap {gap:+d}  ->  P(chosen wins) = {sigmoid(gap):.2f}")
```

## Step 3 — So what should "training" do?

Make the model's **score for chosen exceed its score for rejected** — i.e. maximize
`σ(score_chosen − score_rejected)`, equivalently **minimize** `−log σ(gap)`. That
`−log σ(gap)` is the loss we'll use.

## Recap

- **sigmoid** turns any number into a probability in `(0,1)`; `σ(0)=0.5`.
- **Bradley-Terry**: `P(chosen wins) = σ(score gap)`.
- Training = widen that gap = minimize `−log σ(gap)`.
